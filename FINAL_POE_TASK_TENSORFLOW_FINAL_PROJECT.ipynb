# **EXPLANATION OF WHAT IMAGE CLASSIFICATION IS.**

Image classification is a fundamental task in computer vision, significantly impacting contemporary technology (www.superannotate.com, n.d.). It entails classifying a complete image by attributing a label derived from preceding training data of annotated images. Image classification fundamentally necessitates an in-depth study at the pixel level, as it examines the constituent elements of an image to accurately anticipate the most appropriate label (www.superannotate.com, n.d.). This approach produces useful insights, facilitating informed decisions across diverse applications, from detecting medical irregularities to improving quality control in manufacturing.

An essential element of picture classification is precise data labeling during the training phase, as erroneous labels might lead to mistakes in the model (John, 2024). To attain superior labeling quality, publically accessible datasets are frequently utilized in model training, guaranteeing consistency and precision (John, 2024). Various image classification methodologies can be employed according to the problem specifications, including binary classification (differentiating between two categories), multiclass classification (managing three or more categories), multilabel classification (permitting multiple labels per image), and hierarchical classification (organizing items within a structured class hierarchy). These approaches tackle a range of real-world situations, from basic binary inquiries to intricate hierarchical classifications, rendering them usable across multiple industries (www.superannotate.com, n.d.).

In reality, picture classification is separate from related tasks such as object identification and object localization, although all contribute to thorough image analysis. Image classification assigns a singular label to the entire image, whereas object localization identifies specific regions or objects inside an image, frequently employing bounding boxes to delineate these areas (John, 2024). Object detection advances this concept by both identifying and categorizing several things inside an image. Collectively, these techniques establish a comprehensive framework in computer vision, providing distinct functionalities to meet various image processing requirements—from general classification to the accurate identification and localization of multiple objects (John, 2024).

````python
# Create ImageDataGenerator instances for training and testing
train_datagen = ImageDataGenerator(
    rescale=1./255, # Rescaling pixel values to [0, 1]
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
#(alonsohdzm25, 2021)
#((alarxx, 2023)
````

This code instantiates the `ImageDataGenerator` class employing several data augmentation methods. Below is a comprehensive analysis of the function of each parameter:

- `rescale=1./255`: Normalizes pixel values from the range [0, 255] to [0, 1].
- `shear_range=0.2`: Implements stochastic shear alterations.
- `zoom_range=0.2`: Randomly enlarges the photos by a maximum of 20%.
- `horizontal_flip=True`: Randomly inverts the photos along the horizontal axis.

These augmentations facilitate the generation of more diverse training data, enhancing the resilience and generalization capacity of the machine learning model (John, 2024).

````python
#create data loaders
train_loader = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)
test_loader = train_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

#(Patel, 2017)
````

# **STEP 5: BUILDING A MODEL**

````python
#Building the Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Input(shape=(128, 128, 3)))
model.add(Conv2D(32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Flatten())
model.add(Dense(units=512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=124, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=len(class_names), activation='softmax'))

#(Patel, 2017)
````

This code creates a convolutional neural network (CNN) for the purpose of picture classification (Analytics Vidhya, 2021). The model architecture comprises convolutional layers for feature extraction, max pooling layers to diminish spatial dimensions, a flatten layer to transform input into a 1D array, dense layers for classification, and dropout layers to mitigate overfitting. The concluding dense layer employs a softmax activation function to provide probabilities for each class, so enabling picture classification based on the acquired characteristics.

````python
#Print the Model summary
model.summary()
````

The model summary reveals that the convolutional neural network (CNN) has multiple layers of convolution, max pooling, and dense layers (Analytics Vidhya, 2021). The convolutional layers, utilizing diverse filter sizes, retrieve features from the input images, whereas max pooling layers diminish the spatial dimensions. The model comprises 190,378 trainable parameters, reflecting its complexity and ability to learn from the data. The thick layers, along with dropout for regularization, classify the images into six groups with a softmax activation function (Analytics Vidhya, 2021). This design seeks to optimize feature extraction and classification accuracy, guaranteeing strong performance on the specified dataset.

````python
#Compile Model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
````

The model is constructed with the Adam optimizer, which is effective for gradient descent optimization. The categorical cross-entropy loss function is defined, suitable for multi-class classification tasks using one-hot encoded labels (Analytics Vidhya, 2021). Furthermore, accuracy is designated as the metric for assessing the model's performance, quantifying the ratio of correctly identified cases. This compilation readies the model for training with the specified configurations.

````python
#Early Stopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
````

This code snippet employs early halting, a method utilized to avert overfitting during training. The training procedure will terminate if the validation loss (`val_loss`) fails to improve for 5 consecutive epochs (`patience=5`). Furthermore, `restore_best_weights=True` guarantees that the model weights return to the optimal state recorded during training (John, 2024). This aids in sustaining the model's optimal performance.

````python
# Train the model
history = model.fit(train_loader, epochs=5, validation_data=test_loader, callbacks=[early_stopping], batch_size=32)
````

The training log illustrates the model's performance across 5 epochs. Initially, the model's performance markedly enhances, demonstrating its learning from the data, with training accuracy increasing from 37.27% to 77.54% and validation accuracy from 58.47% to 74.60%. Simultaneously, the training and validation losses diminish, indicating improved model fitting to the data (Analytics Vidhya, 2021). This approach aims to train the CNN for successful image classification while evaluating its performance on unseen validation data to ensure generalization and prevent overfitting.

````python
# Plot training and validation accuracy
plt.plot(history.history['accuracy'],
label='Training Accuracy')
plt.plot(history.history['val_accuracy'],
label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````

This package integrates Keras models with scikit-learn and is useful for managing machine learning workflows. Once installed, we can use SciKerasClassifier in the pipelines similarly to how we use KerasClassifier (John,2024).

````python
import tensorflow as tf
from tensorflow import keras
from scikeras.wrappers import KerasClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import StandardScaler
import numpy as np
# Import to_categorical
from tensorflow.keras.utils import to_categorical

# Create data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_loader = train_datagen.flow_from_directory(
    '/content/seg_train/seg_train',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical' # This ensures labels are one-hot encoded
)

test_loader = test_datagen.flow_from_directory(
    '/content/seg_test/seg_test',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical' # This ensures labels are one-hot encoded
)

# Create a function for the Keras model
def create_keras_model():
    model = keras.Sequential([
        keras.layers.Input(shape=(128, 128, 3)),
        keras.layers.Conv2D(32, kernel_size=(5, 5), activation='relu'),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Conv2D(32, kernel_size=(5, 5), activation='relu'),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        keras.layers.MaxPooling2D(2, 2),
        keras.layers.Flatten(),
        keras.layers.Dense(units=512, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(units=124, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(units=6, activation='softmax')  # Adjust number of classes as needed
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Define EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Convert the data loaders into arrays for compatibility with scikit-learn
x_train, y_train = next(iter(train_loader))
x_test, y_test = next(iter(test_loader))

#create pipeline
pipeline = Pipeline([
    ('keras_model', KerasClassifier(model=create_keras_model, epochs=60, batch_size=32, callbacks=[early_stopping]))
])

#Fit pipeline
pipeline.fit(x_train, y_train)
````

This code establishes a machine learning pipeline utilizing TensorFlow, Keras, and scikit-learn to construct, train, and assess a convolutional neural network (CNN) for image categorization. Initially, data generators are established utilizing `ImageDataGenerator` to preprocess and enhance training and testing images from designated directories. The `create_keras_model` function constructs a CNN model with several convolutional and pooling layers, succeeded by dense layers, then compiles it utilizing the Adam optimizer and categorical cross-entropy loss (Analytics Vidhya, 2021). Early stopping is set to terminate training if the validation loss fails to improve. The training and test data loads are transformed into arrays for compliance with scikit-learn. A pipeline is established and configured with `KerasClassifier`, incorporating the Keras model and specifying training settings including epochs, batch size, and early termination. The pipeline is ultimately applied to the training data to train the model. This systematic method utilizes the advantages of both deep learning and conventional machine learning frameworks for efficient model training and assessment (Analytics Vidhya, 2021).

**Accuracy:** - **Initial Epochs (1-20):** The accuracy begins at a low level but demonstrates a steady enhancement. Initially, the accuracy is approximately 18.75% and increases to 53.12% by epoch 20.
- **Mid Epochs (21-40)**: Accuracy persists in its ascent, attaining around 81.25% by epoch 40. This phase demonstrates greater stability and notable enhancements.

### **3. Recommendations for Improvement**

#### **a. Data Augmentation and Preprocessing**
1. **Enhanced Data Augmentation**:
   - Apply transformations like random cropping, rotation, scaling, flipping, and color jittering.
   - Use techniques like **MixUp** or **CutMix**, which blend or mask parts of images to improve model robustness.

2. **Class Balancing**:
   - Use oversampling techniques like SMOTE or class-weight adjustments during training to balance underrepresented classes.
   - Collect additional images for underrepresented or complex categories.

3. **Data Preprocessing**:
   - Normalize the images and resize them consistently.
   - Perform feature standardization to improve training convergence.

#### **b. Model Architecture and Transfer Learning**
1. **Pretrained Models**:
   - Use architectures like **VGG16**, **ResNet50**, or **EfficientNet** with pretrained weights (e.g., on ImageNet). Fine-tune these models to adapt them to the specific dataset.

2. **Custom Model Enhancements**:
   - Add **batch normalization layers** to stabilize training and improve convergence.
   - Introduce **dropout layers** to reduce overfitting.
   - Increase depth with **residual connections** for better feature extraction.

3. **Vision Transformers (ViTs)**:
   - Explore **ViTs**, which leverage attention mechanisms for robust classification in multi-class image datasets.

#### **c. Training Strategies**
1. **Hyperparameter Optimization**:
   - Use tools like **Optuna** or **Grid Search** to fine-tune learning rates, batch sizes, and optimizer types.

2. **Learning Rate Scheduling**:
   - Implement learning rate schedulers (e.g., ReduceLROnPlateau) to adapt learning rates dynamically.

3. **Early Stopping and Checkpoints**:
   - Adjust early stopping patience to allow more epochs for complex patterns.
   - Save the best model weights to prevent loss of performance during training (Analytics Vidhya, 2021).

#### **d. Regularization Techniques**
1. **Dropout and L2 Regularization**:
   - Increase robustness and prevent overfitting by applying dropout (e.g., 0.2–0.5) and L2 regularization.

2. **Data-Specific Regularization**:
   - Use **label smoothing** to prevent the model from being overly confident in its predictions (Analytics Vidhya, 2021).
